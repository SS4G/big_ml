```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

## seaborn 画图思想
- seaborn 和matplotlib的区别在于 seaborn 更倾向于快速的画图 对 matplot lib中大量的精细的函数直接做了组合
- 可以直接使用pd.DataFrame中的不同的列对数据进行不同维度上的分组

### realplot 点图 (scatter plot)
试图使用不同的图形特征来在一张图中表示多个维度的数据
参数: 
1. x 作为x轴的列 
2. y 作为y轴的列
3. hue 颜色分组 使用hue指定的列将数据分为多组
4. style 点的风格分组
5. size 点的大小分组 
6. col 使用subplot多图模式的分组列
7. row 使用subplot多图模式的分组行

```python
sns.set()
tips = sns.load_dataset("tips")
tips.head(3)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
# 使用多个维度 画出多个特征的组合图
sns.relplot(x="total_bill", y="tip", hue='sex', style='smoker', size='size', row="time", col="day", data=tips)
```

### realplot 线图(line plot)
使用线使用不同的图形特征来在一张图中表示多个维度的数据
参数: 
1. x 作为x轴的列 
2. y 作为y轴的列
3. hue 颜色分组 使用hue指定的列将数据分为多组
4. style 线的风格分组
5. size 线的粗细分组 
6. col 使用subplot多图模式的分组列
7. row 使用subplot多图模式的分组行

```python
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate() #防止x坐标轴的标注重叠 所以对结果进行旋转
```

```python
# 画line图的时候 一般横轴x是时间 所以会默认对 (x,y)数据对按照x从小到大的顺序进行排序
dt = np.random.randn(2000, 2)
df = pd.DataFrame({"x": dt[:, 0], "y": dt[:, 1]})
# 不排序画出的上图 可以认为是一个路径的图像 matplotlib 默认是 不排序的
g = sns.relplot(x="x", y="y", kind="line", sort=False, data=df) 

# 排序画出的下图 可以认为是一个时间的图像 
g = sns.relplot(x="x", y="y", kind="line", data=df)
```

#### 线图 画出聚合值和误差范围
如果在同一时间有多个观测数据 那么可以画出她的平均值 最大值 最小值等

```python
fmri = sns.load_dataset("fmri")
fmri.head(3)
```

```python
# 默认带有置信区间
# 使用不同的ci 关键词参数可以操作置信区间
# ci="sd" standard deviation 标准差
# ci=None 不显示任何区间
# ci 默认为95%置信区间
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
# 使用不同的列对数据进行不同层次维度上的分割 表现数据的多个维度
# 默认使用虚线实线的方案
sns.relplot(x="timepoint", y="signal", hue='region', style='event', kind="line", data=fmri)
```

```python
# 使用不同的列对数据进行不同层次维度上的分割 表现数据的多个维度
# dashes=False, markers=True 表示不使用虚线实线 而是使用点的形状来区分 style
sns.relplot(x="timepoint", y="signal", hue='region', style='event', dashes=False, markers=True, kind="line", data=fmri)
```

- 也存在 hue当前的颜色 表示在数据的线性空间上是不够用的 可能需要将颜色的变化映射到数据的对数尺度上

### 画出类目变量之间的关系
- 在同一个图中 横坐标 一般是类别类型 纵坐标一遍是连续的
- 其他类目信息使用颜色 标点大小来分辨
- 可以使用提琴图(kde)曲线来表示 类别内部的概率密度分布
- 可以使用箱型图来粗略观察数据的分布
- 点形图用来查看不同类目下数据的变化

```python
# cat scatter 可以明显的看出组间的数据分布差异
sns.catplot(x="Survived", y="Age", hue="Sex", kind="swarm", data=train_df)
```

```python
# scatter 
sns.catplot(x="Survived", y="Age", hue="Sex", data=train_df)
```

```python
# kde 提琴图 split参数是画成各一半 只能反映组内数据的概率分布 对于组间的 数据关系 无法反映
sns.catplot(x="Survived", y="Age", hue="Sex", kind="violin", split=True, data=train_df)
```

```python
# bar 默认是计算了对应组数据的平均值 
train_df.query("Sex == 'male'").Age.mean()
sns.catplot(x="Survived", y="Age", hue="Sex", kind="bar", data=train_df)
```

```python
# 组合画图
g = sns.catplot(x="Survived", y="Age", hue="Sex", kind="violin", data=train_df)
n = sns.catplot(x="Survived", y="Age", hue="Sex", color='k', data=train_df, ax=g.ax) # 下面多出来的图怎么消。。
```

```python
sns.catplot(x="Sex", kind="count", hue="Survived", data=train_df);
```

```python
sns.catplot(y="Survived", x="Sex", kind="point", hue="Pclass", data=train_df);
```

```python
npa = np.concatenate([np.random.randn(1000), np.random.rand(1000), np.random.exponential(0.5, size=1000)], axis=0)
npb = np.concatenate([["Norm",] * 1000, ["Uniform",] * 1000, ["Exp",] * 1000], axis=0)
rand_df = pd.DataFrame({"rand_val": npa, "rand_type": npb})
```

```python
rand_df.head(3)
```

```python
# 蓝色 正态分布 
# 橙色 均匀分布 
# 绿色 指数分布
sns.catplot(y="rand_val", x="rand_type", kind="violin", data=rand_df)
```

```python
# 可以看到指数分布中大量的长尾离群点
sns.catplot(y="rand_val", x="rand_type", kind="box", data=rand_df)
```

### 画出单变量的分布 
- 使用distplot 
- 默认kde曲线 直方图 是打开的 可以设置kde=False

```python
sns.distplot(rand_df.query("rand_type == 'Norm'").rand_val)
```

```python
# bins 参数设置 分桶的数量  
sns.distplot(rand_df.query("rand_type == 'Uniform'").rand_val, bins=100, kde=False)
```

```python
# hist 设置是否显示直方图 
# rug 参数的疏密程度反映了 数据分布的位置
sns.distplot(rand_df.query("rand_type == 'Exp'").rand_val, hist=False, rug=True)
```

```python
### 用来查看 两个变量的联合分布
- 使用 jointplot 
- 和relplot 不同之处在于 relplot实际上是个二维的 jointplot 是有深度的 是三维的
```

```python
npa = np.concatenate([np.random.randn(1000), np.random.randn(1000) * 2 + 1], axis=0)
npb = np.concatenate([["std",] * 1000, ["no_std",] * 1000], axis=0)
rand_df = pd.DataFrame({"rand_val": npa, "rand_type": npb})
x_val = rand_df.query("rand_type == 'std'").reset_index().rand_val
y_val = rand_df.query("rand_type == 'no_std'").reset_index().rand_val
joint_df = pd.DataFrame({"X1": x_val, "X2": y_val})
sns.jointplot(x="X1", y="X2", data=joint_df)
```

```python
# 使用二维变量的kde曲线 看起来近似于二维的 正态分布概率密度图
sns.jointplot(x="X1", y="X2", kind='kde', data=joint_df)
```

```python
# 使用一个密度更高的色谱 来画 n_levels 是用来调整 
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.jointplot(x="X1", y="X2", kind='kde', cmap=cmap, n_levels=60, data=joint_df)
```

### 热力图
- 在方格内绘制出对应位置的值 并设置注释
- 其实就是把一个二维表格的数据用颜色 表示出来了

```python
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")
# 绘制x-y-z的热力图，比如 年-月-销量 的热力图
f, ax = plt.subplots(figsize=(9, 6))
#绘制热力图，还要将数值写到热力图上
#每个网格上用线隔开
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
```
